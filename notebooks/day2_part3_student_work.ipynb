```python
# prompt: Let's get import needed Python libraries and download data from the url: "http://peeter.eki.ee:5000/valence/exportparagraphs" and save the result as CSV file. Then load the CSV file into dataframe.

import pandas as pd
import requests

# Download data from the URL
url = "http://peeter.eki.ee:5000/valence/exportparagraphs"
response = requests.get(url)

# Save the result as a CSV file
with open("valence_data.csv", "wb") as f:
  f.write(response.content)

# Load the CSV file into a dataframe
df = pd.read_csv("valence_data.csv")

# head
df.head()
```

## TF-IDF embedding

```python
# let's create TF-IDF embedding for paragraph column 
# and save it into tf_idf column in the dataframe
MAX_TERMS = 2000
from sklearn.feature_extraction.text import TfidfVectorizer

# Initialize TfidfVectorizer
vectorizer = TfidfVectorizer(max_features=MAX_TERMS)

# Fit and transform the 'paragraph' column
tf_idf = vectorizer.fit_transform(df['paragraph'])

# Create a new column 'tf_idf' in the DataFrame
df['tf_idf'] = list(tf_idf.toarray())

# head
df.head()


```

```python
# get dictionary of feature names and their idf values
feature_names = vectorizer.get_feature_names_out()
idf_values = vectorizer.idf_
# print first 5
print("Feature names and their idf values:")
for i in range(5):
  print(f"{feature_names[i]}: {idf_values[i]}")
# last 5
print("Feature names and their idf values:")
for i in range(-5, 0):
  print(f"{feature_names[i]}: {idf_values[i]}")
```

```python
# find which feature has highest idf value
max_idf = max(idf_values)
max_idf_index = idf_values.tolist().index(max_idf)
max_idf_feature = feature_names[max_idf_index]
print(f"Feature with highest idf value: {max_idf_feature} with idf value: {max_idf}")
```

```python
# find what is the lowest idf value
min_idf = min(idf_values)
min_idf_index = idf_values.tolist().index(min_idf)
min_idf_feature = feature_names[min_idf_index]
print(f"Feature with lowest idf value: {min_idf_feature} with idf value: {min_idf}")
```

```python
# find what is the average idf value
avg_idf = sum(idf_values) / len(idf_values)
print(f"Average idf value: {avg_idf}")
```

## Doc2Vec embedding

```python
# now let's create a doc 2 vec embedding for paragraph column
# and save it into doc2vec column in the dataframe
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
from nltk.tokenize import word_tokenize
import nltk

# download punkt
nltk.download('punkt')

```

```python

# Tokenize the paragraphs
tagged_data = [TaggedDocument(words=word_tokenize(_d.lower()), tags=[str(i)]) for i, _d in enumerate(df['paragraph'])]

# Initialize Doc2Vec model
model = Doc2Vec(vector_size=100, window=2, min_count=1, workers=4, epochs=10)

# Build the vocabulary
model.build_vocab(tagged_data)

# Train the Doc2Vec model
model.train(tagged_data, total_examples=model.corpus_count, epochs=model.epochs)

# Get the embeddings
doc2vec = [model.infer_vector(word_tokenize(_d.lower())) for _d in df['paragraph']]
# Create a new column 'doc2vec' in the DataFrame
df['doc2vec'] = doc2vec

# head
df.head()
```

```python
# how many unique values in valence column
print(f"Unique values in 'valence' column: {df['valence'].unique()}")
# so count
print(f"Count of unique values in 'valence' column: {df['valence'].nunique()}")
```

```python
# cluster the embeddings in doc2vec column using kmeans
from sklearn.cluster import KMeans

num_clusters = df['valence'].nunique()
print(f"Number of clusters: {num_clusters} because we have {num_clusters} unique values in 'valence' column")

# Initialize KMeans
kmeans = KMeans(n_clusters=num_clusters, random_state=0)

# Fit the 'doc2vec' column
clusters = kmeans.fit_predict(df['doc2vec'].tolist())

# Create a new column 'cluster' in the DataFrame
df['cluster'] = clusters

# head
df.head()
```

```python
# print value counts of cluster column
df['cluster'].value_counts()
```

```python
# group by valence column and print value counts of cluster column
valence_cluster_counts = df.groupby('valence')['cluster'].value_counts()
valence_cluster_counts
```

```python
# let's plot a correlation matrix valence and cluster columns
# using heatmap from seaborn
import seaborn as sns
import matplotlib.pyplot as plt

# Create a correlation matrix
unique_valences = df['valence'].unique()
# first create numerical representation of valence column category in valence_num col
df['valence_num'] = df['valence'].apply(lambda x: unique_valences.tolist().index(x  ))
correlation_matrix = df[['valence_num', 'cluster']].corr()

# Create a heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')

# Show the plot
plt.show()
```
